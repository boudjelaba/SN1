## Animation de l'affichage des résultats
### Exemple 1 :

```python
import sys
import time
from IPython.display import display, clear_output
for i in range(10):
    time.sleep(0.25)
    clear_output(wait=True)
    print(i)
    sys.stdout.flush()
```

## Signal sinusoïdal sur une période
### Exemple 3 :

```python
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'

fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()

i = 0
t = np.linspace(0,2*np.pi,100)
x, y = [], []

while i<100:
    x.append(t[i])
    y.append(np.cos(t[i]))
    
    ax.plot(x, y, color='r')
    
    fig.canvas.draw()  
    ax.set_xlim(0, t.max())
    ax.set_ylim(-1.1,1.1)
    
    time.sleep(0.1)
    i += 1

ax.grid()
```

## Tracer des courbes animées à partir de données CSV
### Exemple 5 :

```python
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'

import pandas as pd
df = pd.read_csv('bode_diag.csv')
x0 = df['w']
y0 = df['P1']

fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
```

```python
i = 0
x, y = [], []

while i<len(x0):
    x.append(x0[i])
    y.append(y0[i])
    
    ax.semilogx(x, y, color='b', label='Phase')
    
    fig.canvas.draw()
    
    ax.set_xlim(x0.min(), x0.max())
    ax.set_ylim(y0.min(),y0.max())
    
    time.sleep(0.1)
    i += 1

ax.grid()
```
